# Summary 

This content demonstrates how basic a systematic investing strategy can be. 

Individual investors and speculators, in general, would like to believe that the investment decisions they make will be profitable.
Many investors would also like to feel a greater level of "certainty" around the market they are participating in.
A systematic investing strategy that is tested on historical data can provide a more certainty to individuals, even if it merely incremental.

The systematic trading strategy presented here is based on a technical indicator called Bollinger Bands.
Bollinger Bands are a type of statistical chart characterizing the prices and volatility over time of a financial instrument or commodity, using a formulaic method propounded by John Bollinger in the 1980s.

This strategy is not empirical, but it does rely on economic and probabilistic principles. The strategy is based on the idea that prices tend to revert to the mean over time.
When the price of a stock moves significantly away from its average price, it is likely to revert back to the average price over time. Bollinger Bands are used to identify these significant price movements.
The strategy involves buying the stock when the price moves below the lower Bollinger Band and selling the stock when the price moves above the upper Bollinger Band.

The Historical Price data for Albemarle over the past 200 days is shown above.

The red skwiggly lines above and below the black line are the "Bollinger Bands".
Each band represents the stock price of Albemarle if it deviates from its 20-day moving average price by 2 standard deviations.
In theory, 95% of the price action should occur within these bands when we assume a normal distribution. 

A speculator might trade on this theory.
Thinking that when the price of Albemarle closes below the lower band... it will rise back to it's moving average.
Conversely, when the price of Albemarle closes above the upper band... it will fall back to it's moving average.

This is the basis of the trading strategy we will backtest below.

# Backtesting Mean reversion using Bollinger Bands 

This strategy goes long when the price closes below the lower band, and closes that position when the price reverts to the 20-day SMA.
It goes short when the price closes above the upper band, and closes that position when the price reverts to the 20-day SMA.
If you are trading this strategy, you will not capture extreme price movements between the bands,
but you will capture the mean reversion moves that occur most of the time.

The Backtest results are shown below:

If you traded this strategy for the last 15 years, would you have made 62% on your money, and approx. 3% annualized.
You would have executed 282 trades.
Some of the performance metrics (Calmar Ratio , Sharpe Ratio, Sortino Ratio) suggest that this is a high risk strategy.
Where negative returns are large relative to positive returns.

```python
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import pandas as pd
import bt
#Define Price for Backtesting strategy 1
price = data['Close']

# Create bt_data ONCE with price column
bt_data = pd.DataFrame({
    'ALB': data['Close']  # Price data with ticker name
}).dropna()

import numpy as np
import pandas as pd

bt_data['Signal'] = 0.0  # Start neutral

# Entry/exit logic
long_entry = (bt_data['ALB'] < data['Lower'].reindex(bt_data.index))
long_exit = (bt_data['ALB'] > data['SMA'].reindex(bt_data.index))
short_entry = (bt_data['ALB'] > data['Upper'].reindex(bt_data.index))
short_exit = (bt_data['ALB'] < data['SMA'].reindex(bt_data.index))

# Use a stateful approach to avoid conflicting signals
position = 0  # 1 for long, -1 for short, 0 for flat
signals = []

for i in range(len(bt_data)):
    if position == 0:
        if long_entry.iloc[i]:
            position = 1
        elif short_entry.iloc[i]:
            position = -1
    elif position == 1:
        if long_exit.iloc[i]:
            position = 0
    elif position == -1:
        if short_exit.iloc[i]:
            position = 0
    signals.append(position)

bt_data['Signal'] = signals

# Separate the price and signal columns for clarity
price_data = bt_data[['ALB']].copy()
#Create signal DataFrame with SAME column as price
signal_data = pd.DataFrame(index=bt_data.index)
signal_data['ALB'] = bt_data['Signal']

import bt
# Create strategy
strategy1 = bt.Strategy('Bollinger_Bands_1', [bt.algos.RunWeekly(),
    bt.algos.SelectAll(),
    bt.algos.WeighTarget(signal_data),  # Signal column now exists
    bt.algos.Rebalance()
])
# Run backtest
backtest1 = bt.Backtest(strategy1, price_data, initial_capital=10000)
result1 = bt.run(backtest1)
```

# Backtesting Extreme Price moves using Bollinger Bands 

A speculator may know that the equity asset class is, in general, more volatile than fixed income or commodities.
They may use the Bollinger Bands to trade extreme price movements.

A second Bollinger Bands strategy is backtested below.
This strategy goes long when the price closes below the lower band, and only closes that position when the price breaks above the upper band.

It goes short when the price closes above the upper band, and  only closes that position when the price falls back below the lower band.
Aiming to capture the full volatility around the 20 day moving average.

The Backtest results are shown below:

If you had $10,000 and traded this strategy for the last 15 years, would you have made 34% on your money, and less than 2% annualized.
You would have executed 450 trades. Some of the performance metrics (Calmar Ratio , Sharpe Ratio, Sortino Ratio) suggest that this is just as risky as the mean reversion method of trading the bollinger bands.
Simiarly, negative returns here are larger relative to positive returns.

```python
#Define Price for Backtesting strategy
price = data['Close']

# Create bt_data ONCE with price column
bt_data2 = pd.DataFrame({
    'ALB': data['Close']  # Price data with ticker name
}).dropna()

# Add Signal column to SAME DataFrame (don't recreate bt_data!)
bt_data2['Signal'] = 0.0

# Generate Bollinger Band signals
buy_condition = bt_data2['ALB'] < data['Lower'].reindex(bt_data2.index)
sell_condition = bt_data2['ALB'] > data['Upper'].reindex(bt_data2.index)

bt_data2.loc[buy_condition, 'Signal'] = 1.0   # Long
bt_data2.loc[sell_condition, 'Signal'] = -1.0  # Short

# Forward-fill signals to maintain positions
bt_data2['Signal'] = bt_data2['Signal'].ffill().fillna(0.0)

# Separate the price and signal columns for clarity
price_data = bt_data2[['ALB']].copy()

#Create signal DataFrame with SAME column as price
signal_data = pd.DataFrame(index=bt_data2.index)
signal_data['ALB'] = bt_data2['Signal']
```

```python
import bt
# Create strategy
strategy2 = bt.Strategy('Bollinger_Bands_2', [bt.algos.RunWeekly(),
    bt.algos.SelectAll(),
    bt.algos.WeighTarget(signal_data),  # Signal column now exists
    bt.algos.Rebalance()
])

# Run backtest
backtest2 = bt.Backtest(strategy2, price_data, initial_capital=10000)
result2 = bt.run(backtest2)
```
